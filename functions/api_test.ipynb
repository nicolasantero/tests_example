```python
import requests
```

```python
url = f"https://randomuser.me/api/?results=100"
response = requests.get(url, headers={"Accept": "application/json", "Content-Type": "application/json"})
```

```python
if response.status_code == 200:
    data = response.json()

data['results']
```

```python
info = data['results'][0]
info
```

```python
def generate_gender_specific_user(gender, number_of_users):
    url = f'https://randomuser.me/api/?results={number_of_users}&gender={gender}'
    response = requests.get(url)
    return response.json()

generate_gender_specific_user('male', 10)['results']
```
